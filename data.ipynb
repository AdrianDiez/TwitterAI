```python
import pandas as pd

import GLOBALS

with open(GLOBALS.FILE) as json_file:
    df = pd.read_json(json_file)
```

```python
df.info()
```

We need to drop duplicates since the process may create some.

```python
df = df.drop_duplicates(subset=['id'], keep='first')
```

In the feature engineering part we will be cleaning the data and transformimng it to the appropiate way. We want to extract hashtags, links, count values, extract time and date.

```python
df['hashtags'] = [entities.get('hashtags') for entities in df.entities]
df['hashtags'] = [[h.get('text') for h in hashtags] for hashtags in df.hashtags]
```

```python
df['hashtag_num'] = df.hashtags.apply(lambda x: len(x))
```
